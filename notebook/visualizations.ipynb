```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./data/dataset-of-10s.csv")
    
df.drop(["track", "artist", "uri"], axis=1, inplace=True)

display(df.head())
```

```python
def plot_data(data):
    g = sns.PairGrid(data, x_vars=['danceability', 'energy', 'key', 'loudness', 
                                'mode', 'speechiness', 
                                'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature', 'chorus_hit', 'sections'], y_vars='target')    

    try:
        g.map(sns.regplot).set(axis_bgcolor=(1, 1, 1), title='Departure Count vs. Feature')
    except:
        pass

plot_data(df)
```
